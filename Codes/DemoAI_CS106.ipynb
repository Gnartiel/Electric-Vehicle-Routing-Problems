```python
import matplotlib.pyplot as plt

# Read the file
filename = "/content/E-n22-k4.evrp"
with open(filename, 'r') as file:
    data = file.readlines()

# Variables to store the data
coordinates = []
demands = []
stations = []
depot = None

# Parse the data
parse_coords = False
parse_demands = False
parse_stations = False

for line in data:
    if line.startswith("NODE_COORD_SECTION"):
        parse_coords = True
    elif line.startswith("DEMAND_SECTION"):
        parse_coords = False
        parse_demands = True
    elif line.startswith("STATIONS_COORD_SECTION"):
        parse_demands = False
        parse_stations = True
    elif line.startswith("DEPOT_SECTION"):
        parse_stations = False
    elif parse_coords:
        values = line.split()
        if len(values) == 3:
            node_id, x, y = map(int, values)
            coordinates.append((x, y))
    elif parse_demands:
        values = line.split()
        if len(values) == 2:
            customer_id, demand = map(int, values)
            demands.append((customer_id, demand))
    elif parse_stations:
        values = line.split()
        if len(values) == 1:
            station_id = int(values[0])
            stations.append(station_id)
    elif line.strip() == "1":  # DEPOT_SECTION
        depot = 1

# Plotting the data
fig, ax = plt.subplots()

# Plot coordinates as blue dots
x_coords, y_coords = zip(*coordinates)
ax.scatter(x_coords, y_coords, color='blue', label='Customers')

# Plot charging stations as black squares
x_stations = [x_coords[i - 1] for i in stations]
y_stations = [y_coords[i - 1] for i in stations]
ax.scatter(x_stations, y_stations, color='black', marker='s', label='Charging Stations')

# Plot depot as a red square
if depot is not None:
    x_depot, y_depot = x_coords[depot - 1], y_coords[depot - 1]
    ax.scatter(x_depot, y_depot, color='red', marker='s', label='Depot')

# Add labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Plot of Customers, Charging Stations, and Depot')
ax.legend()

# Display the plot
plt.show()
```
